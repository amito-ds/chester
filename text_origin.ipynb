# Imports

```python
import tcap 
from tcap import run_full_cycle
from tcap.data_loader import webtext_data
import pandas as pd
from tcap.run_full_cycle import run_tcap, DataSpec
from tcap.applications.sanity.teacup import find_similar_texts
from tcap.features_engineering.feature_main import FeatureExtraction
from tcap.cleaning.cleaning_func import TextCleaner
from tcap.preprocessing.preprocessing_func import TextPreprocessor
```

# Run 

## Run the full process 

```python
# Get some dataa 
df1 = webtext_data.load_data_pirates().assign(target='pirates')
df2 = webtext_data.load_data_king_arthur().assign(target='king')
df = pd.concat([df1, df2])
df.shape
```

```python
# The outputs include the original dataframe (origin_df), 
# the processed dataframe (df), the embeddings for the training data 
# (train_embedding), the embeddings for the test data (test_embedding), 
# and the trained and fitted model (best_model.model) ready for prediction.
origin_df, df, train_embedding, test_embedding, best_model = \
run_tcap(data_spec=DataSpec(df=df, text_column='text', target_column='target'))
```

```python
best_model.model
```

## Play with run parameters and change the default setting 

You can play with which parts to run, for example if you want only the model without plots, you can leave out 3 chapters. 

if you know you need higher embedding dimension, you can also play with it

```python
# The function's parameters can be adjusted to suit our needs, 
# and cleaning and preprocessing can be performed together
# Apply clean and preprocess together 
# Then Embedding. here you can choose to train a model or not
origin_df, df, train_embedding, test_embedding, best_model = run_tcap(
    data_spec=DataSpec(df=df, text_column='text', target_column='target'),
    is_text_cleaner = True, 
    is_text_preprocesser = True,
    is_text_stats = False, 
    is_feature_extraction = True, 
    is_feature_analysis = False, 
    is_train_model = True, 
    is_model_analysis = False
)
```

## Breakdown the process 

An alternative approach is to optimize each step separately. 

This can lead to a better model with a slightly higher level of effort. 

It is particularly useful for experimenting with different cleaning methods, exploring various embedding techniques and dimensions, and evaluating model performance.

```python
# Cleaning, preprocessing, and text statistics can be run together.
# Experiment with different parameter settings until you find a set that 
# works well for your data. 
# Once you are satisfied with the results, move on to the next step and make sure 
# to review the generated text statistics as it will provide useful information 

text_cleaner = TextCleaner(remove_punctuation_flag = True,
                 remove_numbers_flag = True,
                 remove_whitespace_flag = True,
                 remove_empty_line_flag = True,
                 lowercase_flag = True,
                 remove_stopwords_flag = True,
                 stopwords = None, # you can change stopwords
                 remove_accented_characters_flag = True,
                 remove_special_characters_flag = True,
                 remove_html_tags_flag = True)

text_preprocesser = TextPreprocessor(stemmer=None, 
                                     # you can choose stemmerand lemmatizer 
                 lemmatizer=None,
                 stem_flag=True,
                 lemmatize_flag=False,
                 tokenize_flag=True)

# Text cleaning and preprocessing: play with the parameters 
origin_df, df, train_embedding, test_embedding, best_model = run_tcap(
    data_spec=DataSpec(df=df, text_column = 'text', target_column='target'),
    text_cleaner = text_cleaner, text_preprocesser = text_preprocesser,
               is_text_cleaner = True, 
                is_text_preprocesser = True,
                is_text_stats = True, 
                is_feature_extraction = False, 
                is_feature_analysis = False, 
                is_train_model = False, 
                is_model_analysis = False)
```

```python
# In the next step, you can extract the embeddings and train a model. 
# Experiment with different feature extraction dimensions and model types 
# until you find the best fit for your data

origin_df, df, train_embedding, test_embedding, best_model = run_tcap(
    data_spec=DataSpec(df=df, text_column='text', target_column='target'),
    feature_extraction=FeatureExtraction(split_data=True, corex_dim=80, tfidf_dim=90, bow_dim=100),
    is_text_cleaner=True,
    is_text_preprocesser=True,
    is_text_stats=True,
    is_feature_extraction=True,
    is_feature_analysis=True,
    is_train_model=True,
    is_model_analysis=True)
```
