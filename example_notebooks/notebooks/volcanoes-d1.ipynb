# Welcome to MadCat 🐈
The madcat package provides an end-to-end machine learning modeling process for data analysis. This includes pre-processing, model selection and training, cross-validation, evaluation, and post-model analysis. The package is designed to handle both regression and classification problems, and works with numeric, categorical, boolean, time series, text and images, features.
It can also analyze a video.

```python
import pandas as pd
from sklearn.datasets import fetch_openml
import chester
from chester import run as chester_run
from chester.run import full_run as fr
from chester.run import user_classes as uc
```

# Run MadCat 

# Example

```python
# function to get the data 
def load_dataset(name):
    print("DB name:", name)
    data = fetch_openml(name=name, as_frame=True)
    df = data.data
    df['target'] = data.target
    df = df.dropna(axis=1, how='all')  # remove nas
    return df
```

```python
df = load_dataset('volcanoes-d1')
```

```python
# Run full process
output_collector = fr.run_madcat(uc.Data(df=df, target_column='target'), model_run=uc.ModelRun(n_models=2))
```
